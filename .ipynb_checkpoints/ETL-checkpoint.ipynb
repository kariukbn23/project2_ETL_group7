```python
# Import necessary dependencies for data files to be used. 
import csv 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#After csv from Kaggle was saved to resources folder for this project, we needed to create a file path to access that csv. 
#To load that csv to a df, we performed the pd.read_csv function calling the variable stored above. 
twitter_file = "Resources/nba_2017_twitter_players.csv"

df_twitter_stats = pd.read_csv(twitter_file)
```

```python
#To see if our process above worked we performed a .head() to view our data. 
df_twitter_stats.head()
```

```python
# We repeated this process for our real_plus minus data. Please refer to the steps above. 
RPM_file = "Resources/nba_2017_real_plus_minus.csv"

df_RPM_stats = pd.read_csv(RPM_file)
```

```python
#Refer to previous comment 
df_RPM_stats.head()
```

```python
#Prior to merging these two data sets, we knew that the column ID for the desired merged-on data set, would have to be the same. 
# This is why we performed a .rename(columns={}) to make sure that our second data set matched on the column PLAYER.
df_RPM_stats2 = df_RPM_stats.rename(columns={"NAME": "PLAYER"})

df_RPM_stats2.head()
```

```python
#This is where the real fun began. 
#When we tried performing the pd.merge function on our data sets we continued to get similar errors. 
#When we performed a right merge on the data set, it would return only the player and twitter statistics. 
#When we performed a left merge on the data set, it would only return the real_plus minus data and not the twitter stats. 
#We researched how to perform merges on at least 3 files because we thought we could join our merged df with the data frame that was not showing up. And that didn't work. 
#After brainstorming what was the root issue with our code, we found out that there was an error on our merged-"on" data set. 
#In the original CSV file for our RPM file, the player names also included the players basketball position. 
#We found out that you are unable to perform a successful merge unless both sets of data on the selected merge-"on" criteria are the same. 
#In other words since our twitter stats only had the player name we needed to make sure that our RPM data set only had the player name as well. 
#After we hard edited the RPM csv to not include the position of the player, we then reloaded that csv into the file path for our script. And wahlah! All the data in our pd.merge finally showed up! 
#See below :) 
merged_df = pd.merge(df_RPM_stats2, df_twitter_stats, how = "left", on = "PLAYER")

merged_df.head()
```

```python
#After getting our merged data set using the two different csv files, we then cleaned up the df, by eliminating the column values we did not need. 
final_df1 = merged_df[["PLAYER", "RPM", "TWITTER_FAVORITE_COUNT"]]

final_df1.head()
```

```python
#While this was not a requirement of the project, we wanted to get a visual to help us answer our essential question to find out whether or not a correlation existed between a player's popularity and how well they performed for their respective team. 
#What we found was that there was no correlation between how many times a player's tweet was favorited and how they performed on the court. So no surprises there. 
plt.figure(figsize=(8,6))
plt.grid
plt.scatter(final_df1["TWITTER_FAVORITE_COUNT"],final_df1["RPM"],  edgecolor = "black", color="blue", alpha=1)
plt.ylabel("RPM")
plt.xlabel("TWITTER_FAVORITE_COUNT")
plt.show()
```

```python
#For our second research question, we followed the same logical progression as the first half of this script. Please refer to the comments in the lines above for the rationale used on the steps below. 
salary_file = "Resources/nba_2017_salary.csv"

df_salary_info = pd.read_csv(salary_file)
```

```python
df_salary_info.head()
```

```python
df_salary_info = df_salary_info.rename(columns={"NAME": "PLAYER"})

df_salary_info.head()
```

```python
new_final_df = pd.merge(df_salary_info, final_df1, on = "PLAYER", how = "left")

new_final_df.head()
```

```python
new_final_df2 = new_final_df[["PLAYER", "SALARY", "RPM"]]

new_final_df2.head()
```

```python
#Here is where it got fun again! 
#The big moment of truth! 
#As seen below, we were able to land on there actually being a positive correlation between how much players were paid that season and their overall production on the basketball court. 
#For example, it makes sense why LeBron James and Steph Curry were the two highest paid players in the league that year and they both also had some of the highest real_plus_minus (RPM) statistics in the league. 
#For those unfamiliar with RPM, it is a player's average impact in terms of net point differential per 100 offensive and defensive possessions. 
plt.figure(figsize=(8,6))
plt.grid
plt.scatter(new_final_df2["SALARY"],new_final_df2["RPM"],  edgecolor = "black", color="blue", alpha=1)
plt.ylabel("RPM")
plt.xlabel("Salary")
plt.show()
```
